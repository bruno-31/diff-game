```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.load('sgd.npy')
y=np.load('concensus.npy')
z=np.load('kl_l2_1e4.npy')
w=np.load('kl_l2_1e3.npy')
```

```python
t=np.linspace(0,20000,len(x))
plt.semilogy(t,w)
plt.semilogy(t,y)
plt.semilogy(t,z)
plt.semilogy(t,x)

plt.ylabel('kl divergence')
plt.legend(['sgd', 'concensus','l2 lambda = 1e-4', 'l2 lambda = 1e-3'])
fig = plt.gcf()
fig.savefig('./kl.png', dpi=200)
```
