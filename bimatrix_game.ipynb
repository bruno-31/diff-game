```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def nabla_x(A,y):
    return np.matmul(A,y) 

def nabla_y(A,x):
    return -np.matmul(A.T,x) 

def loss(x,A,y):
    return np.squeeze(np.matmul(x.T,np.matmul(A,y)))
```

## Simultaneous gradient descent

```python
n = 100
_lambda = 0.05
_eta = 0.01
A = np.random.random(size=[n,n])
x = np.random.random(size=[n,1])
y = np.random.random(size=[n,1])
```

```python
l1 = [];
for i in range(600):
    x -= _eta * nabla_x(A,y)
    y -= _eta * nabla_y(A,x)
    l1.append(loss(x,A,y))
plt.plot(l1)
plt.show()    
```

## tf.contrib.GeorgiosOptimizer

```python
n = 2
_lambda = 0.1
_eta = 0.01
#A = np.random.random(size=[n,n])
A = np.array([[1, -1], [-1, 1]])
x = np.random.random(size=[n,1])
y = np.random.random(size=[n,1])
```

```python
steps = 600
l1 = [];_x = np.empty((steps, 2));
for i in range(steps):
    x -= _eta * (nabla_x(A,y) + _lambda * x)
    y -= _eta * (nabla_y(A,x) + _lambda * y)
    l1.append(loss(x,A,y))
    _x[i, :] = x.squeeze()
    print(x, y, loss(x, A, y))

plt.plot(l1)
plt.show()    
```

## Matching Penny 

```python
A = np.array([[1,-1],[-1,1]])
x = np.array([1.,0.])
y = np.array([0,1.])
_lambda = 0.3
_eta = 0.05
```

```python
steps = 600
l1 = [];_x = np.empty((steps, 2));_y = np.empty((steps, 2));
for i in range(steps):
    x -= _eta * (nabla_x(A,y)+ _lambda * x)
    y -= _eta * (nabla_y(A,x)+ _lambda * y)
    l1.append(loss(x,A,y))
    _x[i, :] = x.squeeze()
    _y[i, :] = y.squeeze()
#     print(x, y, loss(x, A, y))
plt.figure(1,figsize=(18,5))
plt.subplot(121)
plt.plot(l1)
plt.title('loss')
plt.subplot(122)
plt.scatter(_x[:,0],_y[:,0],c=np.linspace(0.,0.01,steps),s=5,cmap='hot')
plt.xlabel('x')
plt.ylabel('y')

plt.show()    
```

```python
steps = 600
nabla_x2 = 0
nabla_y2= 0
l1 = [];_x = np.empty((steps, 2));_y = np.empty((steps, 2));
for i in range(steps):
    nabla_x2 = nabla_x(A,y)
    nabla_y2 = nabla_y(A,x)
    x += 2*_eta * nabla_x(A,y) - _eta * nabla_x2
    y += - 2*_eta * nabla_y(A,x) + _eta * nabla_y2
    
    l1.append(loss(x,A,y))
    _x[i, :] = x.squeeze()
    _y[i, :] = y.squeeze()
#     print(x, y, loss(x, A, y))
plt.figure(1,figsize=(18,5))
plt.subplot(121)
plt.plot(l1)
plt.subplot(122)
plt.scatter(_x[:,0],_y[:,0],c=np.linspace(0.,0.01,steps),s=5,cmap='hot')
plt.show()    
```
