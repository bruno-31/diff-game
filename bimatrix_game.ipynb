```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def nabla_x(A,y):
    return np.matmul(A,y) 

def nabla_y(A,x):
    return -np.matmul(A.T,x) 

def loss(x,A,y):
    return np.squeeze(np.matmul(x.T,np.matmul(A,y)))
```

## Matching Penny 

```python
## SGD (+l2) 
```

```python
A = np.array([[1,-1],[-1,1]])
x = np.array([1.,0])
y = np.array([1.,0])
_lambda = 0.5
_eta = 0.008

steps = 2000
l1 = [];_x = np.empty((steps, 2));_y = np.empty((steps, 2));
for i in range(steps):
    x -= _eta * (nabla_x(A,y)+ _lambda * x)
    y -= _eta * (nabla_y(A,x)+ _lambda * y)
    l1.append(loss(x,A,y))
    _x[i, :] = x.squeeze()
    _y[i, :] = y.squeeze()
#     print(x, y, loss(x, A, y))
plt.figure(1,figsize=(10,5))
plt.subplot(121)
plt.plot(l1)
plt.subplot(122)
plt.scatter(_x[:,0],_y[:,0],c=np.linspace(0.,0.01,steps),s=5,cmap='hot')
plt.xlabel('x')
plt.ylabel('y')

plt.show()    
```

```python
_x0 = _x
_y0 = _y
_l0 = l1
#0
```

```python
_x1 = _x
_y1 = _y
_l1 = l1
#0.2
```
